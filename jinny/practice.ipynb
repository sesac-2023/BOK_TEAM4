```python
data = df.groupby('date').agg({'title': '/'.join, 'contents': '/'.join}).reset_index()
```

```python
melted_data = pd.melt(data, id_vars=['date'], value_vars=['title', 'contents'], var_name='column', value_name='title_contents')
```

```python
result = melted_data.groupby('date')['title_contents'].agg('&'.join).reset_index()
```
