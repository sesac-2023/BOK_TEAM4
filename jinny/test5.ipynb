```python
import pandas as pd
```

```python
df = pd.read_csv('./news_1000.csv', index_col=0)
```

```python
word_list = []
```

```python
for i in range(0, 1000):
  target = df.iloc[i]
  info_1 = target["text"]
  info_2 = target["up_down"]
  word_list.append([info_1, info_2])
```

```python
MPB_df = pd.read_csv('./minutes_call.csv', index_col=0)
```

```python
MPB_df
```

```python
target_text = MPB_df["text"][MPB_df['date'] == '2009-05-21']
```

```python
df_text = target_text.to_string(index=False)
```

```python
df_text
```

```python
word_list
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from ekonlpy import Mecab
#모델 훈련 및 결과출력
# def Model_test(word_list,file_text):
    # 1. 데이터 준비
X_train = [i[0] for i in word_list]
y_train = [i[1] for i in word_list]
# 2. 특성 추출 (형태소 분석 및 n-그램)
mecab = Mecab()
tagger = mecab.pos
mecab_tokenize = lambda text: [token for token, pos in tagger(text)  if pos == "NNG" or pos == "VA" or pos == "VAX" or pos == "MAG" or pos == "VV"]
vectorizer = CountVectorizer(tokenizer=mecab_tokenize, ngram_range=(1, 5), min_df=15)
X_train_counts = vectorizer.fit_transform(X_train)
# 3. 나이브 베이즈 모델 학습
clf = MultinomialNB()
clf.fit(X_train_counts, y_train)
print(clf.score(X_train_counts, y_train))
#4. 새로운 입력 예측
X_test = [df_text]
X_test_counts = vectorizer.transform(X_test)
predicted_labels = clf.predict(X_test_counts)
print(predicted_labels)
```
