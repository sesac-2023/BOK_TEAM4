```python
def title(text):    
    text = re.sub(r'<[^>]+>', '', text)  # HTML 태그 제거
    text = text.replace('\"', '').replace('\'', '')  # 따옴표 제거
    text = re.sub(r'[\[\]\(\)▶◀#$^&*[\]{}<>/|』◇◆▲△■□=·●]', '', text)  # 특수기호 및 특수문자 제거
    return text.strip()
```

```python
def clean_content(text):
    text = re.sub(r'<[^>]+>', '', text)  # HTML 태그 제거
    text = text.replace('\"', '').replace('\'', '')  # 따옴표 제거
    text = re.sub(r'[\[\]\(\)▶◀#$^&*[\]{}<>/|』◇◆▲△■□=·●]', '', text)  # 특수기호 및 특수문자 제거
    text = text.replace('\n', '').replace('\r', '')  # 줄바꿈 문자 제거
    text = re.sub(r'\S+@\S+', '', text)  # 이메일 주소 제거
    return text.strip()
```
