```python
for i in range(len(contents_list)):
    for target in replace_targets:
        contents_list[i] = contents_list[i].replace(target, ' ')
```

```python
#뉴스 이름 제거
company_name = news['company'].unique()
for i in range (len(contents_list)):
  for name in company_name:
      contents_list[i] = contents_list[i].replace(name, ' ')
```

```python
#특수기호 제거
special_chars = ['☞', 'ⓒ','※', '★']
for i in range(len(contents_list)):
  for char in special_chars:
     contents_list[i] = contents_list[i].replace(char, ' ')
```

```python
#기자 이름 제거
contents_list = [re.sub(r'\w+\s?\w*?\s?기자', '', content) for content in contents_list]
```

```python
# 'id@' , @id제거제거
contents_list = [re.sub(r'\b[a-z0-9]+@', '', content) for content in contents_list]
contents_list = [re.sub(r'@[a-z0-9]+', '', content) for content in contents_list]
```

```python
#url 제거
contents_list = [re.sub(r'(www\.)?\w+(\.\w+){2,}', '', content) for content in contents_list]
```

```python
#연속된 공백을 하나의 공백으로 바꾸기
contents_list = [re.sub(r'\s+', ' ', content) for content in contents_list]
```
